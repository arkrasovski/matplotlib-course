```python
x = np.arange(100)

fig, axs=plt.subplots(2,2) #2 row 2 column

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine Wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine Wave")

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title("Random func")

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title("Log func")

axs[1, 1].set_xlabel('TEST')

plt.tight_layout() #    
fig.suptitle('Four plots')
```

```python
plt.figure()
ax = plt.axes(projection='3d')
x = np.arange(0, 50, 0.1)
y = np.arange(0, 50, 0.1)
z = np.cos(x+y)

ax.plot(x,y,z)
ax.set_title('3d plot')
ax.set_xlabel('test')
plt.show()
```

```python
plt.figure()
ax = plt.axes(projection='3d')
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y) #cross join

Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap='Spectral')
ax.set_title('3d plot')
ax.set_xlabel('test')

ax.view_init(azim=0, elev=90)# left right = azimov, up down = elevation

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation


#  
heads_tails = [0, 0]

#    
fig, ax = plt.subplots()
bar_container = ax.bar(['Heads', 'Tails'], heads_tails, color=['red', 'blue'])



#    
def update(frame):
    #   
    heads_tails[np.random.randint(0, 2)] += 0.001
    #   
    for bar, height in zip(bar_container, heads_tails):
        bar.set_height(height)
    return bar_container

# 
ani = FuncAnimation(fig, update, frames=range(100), repeat=False)

#  
plt.show()
```
